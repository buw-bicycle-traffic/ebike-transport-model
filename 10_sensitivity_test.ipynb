```python
import os
import win32com.client
import numpy as np
```

```python
verPath = r'C:\Users\arning.FUL\Desktop\ver\DEV_20250102_0830.ver'
Visum = win32com.client.Dispatch('Visum.Visum.24')
Visum.loadVersion(verPath)
```

```python
# Matrix numbers for calculating modal share 
all_cbike = 61
all_ebike = 62
all_fuss = 24
all_pkw = 26
all_pt = 27
all_pkwmitf = 28
```

```python
# Walking
foot_trips_matrix = np.array(Visum.Net.Matrices.ItemByKey(all_fuss).GetValues())
foot_dis_matrix = np.array(Visum.Net.Matrices.ItemByKey(100).GetValues())

Anzfoottrips = foot_trips_matrix.sum()
avfoottriplength = round((foot_trips_matrix * foot_dis_matrix).sum()/Anzfoottrips, 1)
print('Walking av. trip length: '+ str(avfoottriplength))
print('Modal split:')
print(np.array(Visum.Net.Matrices.ItemByKey(all_fuss).GetValues()).sum()/
      (np.array(Visum.Net.Matrices.ItemByKey(all_ebike).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_cbike).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_fuss).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pkw).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pt).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pkwmitf).GetValues()).sum())*100)

# C-Bike
cbike_trips_matrix = np.array(Visum.Net.Matrices.ItemByKey(all_cbike).GetValues())
cbike_dis_matrix = np.array(Visum.Net.Matrices.ItemByKey(5088).GetValues())

Anzcbiketrips = cbike_trips_matrix.sum()
avcbiketriplength = round((cbike_trips_matrix * cbike_dis_matrix).sum()/Anzcbiketrips, 1)
print('C-bike av. trip length: '+ str(avcbiketriplength))
print('Modal split:')
print(np.array(Visum.Net.Matrices.ItemByKey(all_cbike).GetValues()).sum()/
      (np.array(Visum.Net.Matrices.ItemByKey(all_ebike).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_cbike).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_fuss).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pkw).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pt).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pkwmitf).GetValues()).sum())*100)
    
# E-Bike
ebike_trips_matrix = np.array(Visum.Net.Matrices.ItemByKey(all_ebike).GetValues())
ebike_dis_matrix = np.array(Visum.Net.Matrices.ItemByKey(5089).GetValues())

Anzebiketrips = ebike_trips_matrix.sum()
avebiketriplength = round((ebike_trips_matrix * ebike_dis_matrix).sum()/Anzebiketrips, 1)
print('E-bike av. trip length: '+ str(avebiketriplength))
print('Modal split:')
print(np.array(Visum.Net.Matrices.ItemByKey(all_ebike).GetValues()).sum()/
      (np.array(Visum.Net.Matrices.ItemByKey(all_ebike).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_cbike).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_fuss).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pkw).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pt).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pkwmitf).GetValues()).sum())*100)

# car
car_trips_matrix = np.array(Visum.Net.Matrices.ItemByKey(all_pkw).GetValues())
car_dis_matrix = np.array(Visum.Net.Matrices.ItemByKey(100).GetValues())
    
Anzpkwtrips = car_trips_matrix.sum()
avpkwtriplength = round((car_trips_matrix * car_dis_matrix).sum()/Anzpkwtrips, 1)
print('Car (driver) av. trip length: '+ str(avpkwtriplength))
print('Modal split:')
print(np.array(Visum.Net.Matrices.ItemByKey(all_pkw).GetValues()).sum()/
      (np.array(Visum.Net.Matrices.ItemByKey(all_ebike).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_cbike).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_fuss).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pkw).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pt).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pkwmitf).GetValues()).sum())*100)

# car (passenger)
carp_trips_matrix = np.array(Visum.Net.Matrices.ItemByKey(all_pkwmitf).GetValues())
carp_dis_matrix = np.array(Visum.Net.Matrices.ItemByKey(100).GetValues())

Anzpkwmitftrips = carp_trips_matrix.sum()
avpkwmitftriplength = round((carp_trips_matrix * carp_dis_matrix).sum()/Anzpkwmitftrips, 1)
print('Car (passeger) av. trip length: '+ str(avpkwmitftriplength))
print('Modal split:')
print(np.array(Visum.Net.Matrices.ItemByKey(all_pkwmitf).GetValues()).sum()/
      (np.array(Visum.Net.Matrices.ItemByKey(all_ebike).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_cbike).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_fuss).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pkw).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pt).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pkwmitf).GetValues()).sum())*100)

# PT
pt_trips_matrix = np.array(Visum.Net.Matrices.ItemByKey(all_pt).GetValues())
pt_dis_matrix = np.array(Visum.Net.Matrices.ItemByKey(108).GetValues())

Anzpttrips = pt_trips_matrix.sum()
avpttriplength = round((pt_trips_matrix * pt_dis_matrix).sum()/Anzpttrips, 1)
print('PT av. trip length: '+ str(avpttriplength))
print('Modal split:')
print(np.array(Visum.Net.Matrices.ItemByKey(all_pt).GetValues()).sum()/
      (np.array(Visum.Net.Matrices.ItemByKey(all_ebike).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_cbike).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_fuss).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pkw).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pt).GetValues()).sum()+
       np.array(Visum.Net.Matrices.ItemByKey(all_pkwmitf).GetValues()).sum())*100)
```
